Rock Paper Scissors game code in single code cell

```python
import random
from google.colab import userdata

# reading the player name from notebook secrets ( I have assigned PLAYER_NAME as Vidyu)
PLAYER_NAME = userdata.get('PLAYER_NAME')

# displaying welcome message
print("\n***********************************************************************\n")
print(f"Welcome {PLAYER_NAME}, to our Rock-Paper-Scissors Game!")
print("\n***********************************************************************\n")

# valid choices
choices = ["rock", "paper", "scissors"]

# game stats
games_played = 0
games_won = 0
games_tied = 0

# game loop (reference: Claude.ai)
while True:
    # aksing for user choice
    user_choice = input("Please choose an option ('rock', 'paper', or 'scissors'): ").strip().lower()

    # validation loop for input
    while user_choice not in choices:
        print("Invalid choice. Please enter 'rock', 'paper', or 'scissors'.")
        user_choice = input("Try again: ").strip().lower()

    # computer choice
    computer_choice = random.choice(choices)

    # displaying choices
    print("\n-----------------------------------------------------------------------")
    print(f"User choice: '{user_choice}'")
    print(f"Computer choice: '{computer_choice}'")
    print("\n-----------------------------------------------------------------------")

    # winner?
    if user_choice == computer_choice:
        print("It's a tie! :)")
        games_tied += 1
    elif (
        (user_choice == "rock" and computer_choice == "scissors") or
        (user_choice == "paper" and computer_choice == "rock") or
        (user_choice == "scissors" and computer_choice == "paper")
    ):
        print("You won!!! :D")
        games_won += 1
    else:
        print("Computer Won! :(")

    games_played += 1

    # if user wants to play again?
    play_again = input("Would you like to play again? (yes/no): ").strip().lower()
    if play_again not in ["yes", "y"]:
        break

# displaying game stats
print("\n***********************************************************************\n")
print(f"Thanks for playing, {PLAYER_NAME}!")
print(f"Games played: {games_played}")
print(f"Games won: {games_won}")
print(f"Games tied: {games_tied}")
print("\n***********************************************************************\n")

# calculating win% (I'm excluding ties)
if games_played - games_tied > 0:
    win_percentage = (games_won / (games_played - games_tied)) * 100
    print(f"Win percentage (excluding ties): {win_percentage:.2f}%")
else:
    print("Could not calculate win percentage.")
```
